```python
import psycopg2
import sys, os
import numpy as np
import pandas as pd
import pandas.io.sql as psql
```

```python
# Set up a connection to the postgres server.
conn_string = "host="+ PGHOST +" port="+ "5432" +" dbname="+ PGDATABASE +" user=" + PGUSER \
+" password="+ PGPASSWORD
conn=psycopg2.connect(conn_string)
print("Connected!")
```

```python
# Create a cursor object
cursor = conn.cursor()


def load_data(schema, table, skip = 0, limit = 50):

    sql_command = "SELECT * FROM {}.{} OFFSET {} LIMIT {};".format(str(schema), str(table), skip, limit)
    print (sql_command)

    # Load the data
    data = pd.read_sql(sql_command, conn)

    print(data.shape)
    return (data)
```

```python
def get_dataframe(table_name, skip, limit, rename = True):
    df = load_data('public', table_name, skip, limit)
    if rename:
        df = df.rename(columns = FIELD_MAPPER[table_name])
    return df
```

```python
class Person:
    def __init__(self, age, name):
        self.age = age
        self.name = name

    def __eq__(self, other):
        return self.age == other.age and self.name == other.name

    def __hash__(self):
        print('The hash is:')
        return hash((self.age, self.name))

person = Person(23, 'Adam')
print(hash(person))
```
