```python
import pandas as pd  
```

```python
flow_block_df = pd.read_pickle("output/flow-block.pkl")
```

```python
describe_all_episode = []
for i in flow_block_df['block-id'].unique():
    description = {}
    df_descriptions = flow_block_df[flow_block_df['block-id'] == i].describe()
    description['mean_interval-timestamp'] = df_descriptions['interval-timestamp']['mean']
    description['std_interval-timestamp'] = df_descriptions['interval-timestamp']['std']
    description['interval-distribution'] = df_descriptions['reception-timestamp']['max'] - df_descriptions['transmission-timestamp']['min']
    description['block_size'] = df_descriptions['block-size']['mean']
    
    describe_all_episode.append(description)

df_describe = pd.DataFrame(describe_all_episode)
```
